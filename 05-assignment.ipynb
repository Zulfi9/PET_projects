## Задание 3

В этом поможет тип данных колонки. Категориальные переменные сохрани в переменную `X_cat`, а количественные в `X_num`.

```python
X_cat = X.select_dtypes(include = ['object'])
X_num = X.select_dtypes(include = ['int', 'float'])

print ('Число категориальных переменных:', X_cat.shape[-1])
print ('Число количественных переменных:', X_num.shape[-1])

## разделение по типу данных, object - категориалььные (принадлежность к каой-либо категории)
```

## Задание 4
Предобработаем количественные переменные. Проведи стандартизацию для признаков X_num.
Сохрани результат стандартизации в переменную `X_num_scal`.
Выведи среднее и стандартное отклонение по датасету для каждого признака после стандартизации.
Значения округли для сотых.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  ## приводит дисперсию обрабатываемых данных к единице, средние распределение стремится к нулю
X_num_scal = scaler.fit_transform(X_num) # результат стандартизации (чтобы все признаки были примерно в одной шкале)
df_X_num_scal = pd.DataFrame(X_num_scal, columns=X_num.columns.tolist()) # создаем датафрейм
# print(df_X_num_scal)

# df_X_num_scal.shape
```

```python
print('Среднее отклонение после стандартизации:')
df_X_num_scal.mean(axis=0).round(2)

#если среднее отклонение 0, то все ок, стандартизация прошла хорошо
```

```python
print('Стандартное отклонение после стандартизации')
df_X_num_scal.std(axis = 0).round(2)

## Если стандартное отклонение = 1, то все ок, стандартизация проведена коррректно
```

## Задание 5
Алгоритмы машинного обучения плохо работают с категориальными признаками, поэтому их надо трансформировать.
Прочитай про One-Hot Encoding подход к кодированию и примени его с нашим категориальным признаком, сохрани
результат в переменную `X_cat_ohe`
Перед этим удали признак `CLNT_JOB_POSITION` (у него слишком много значений).
Выведи количество колонок таблицы `X_cat_ohe`

```python
# X_cat = X_cat.drop(columns = ['CLNT_JOB_POSITION_TYPE'], axis = 1)   # удаляем признак CLNT_JOB_POSITION
X_cat = pd.DataFrame(X_cat.drop('CLNT_JOB_POSITION', axis=1))
X_cat.shape
# X_cat
```

```python

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown = 'ignore')  # некатегориальные признаки считаются нулями, возвращает массивы, не датафрейм
X_cat_ohe = encoder.fit_transform(X_cat).toarray() # из длинных данных делает широкие (смотрим сколько уникальных значений в столбе, данные строковые, переводит в столбцы, индекс столбца - количество встреч, количество строк отстается неизменным)
print('Количество колонок таблицы X_cat_ohe =', X_cat_ohe.shape[-1])
```

```python
# X_cat_ohe
```

```python
df_X_cat_ohe = pd.DataFrame(X_cat_ohe)
df_X_cat_ohe.columns = encoder.get_feature_names_out()  ## Метод для получения новых колонок склеивает название признака и его значение
df_X_cat_ohe
```

## Задание 6
Объедини 2 предобработанных датасета. Результат запиши в переменную `X_ready`. Сохрани датасет в папку `datasets/data_prepared.csv`
Подели получившийся датасет на `train` и `test`. Процент тестовой выборки: `20%`. Использовать параметр `random_state=21`.
Выведи размерность обучающей выборки X_train.


```python
## объединяем датасеты
X_ready = pd.concat([df_X_num_scal, df_X_cat_ohe], axis=1) ## объединение по столбцам
X_ready.shape
```

```python
## Сохранение датасета в папку
from pathlib import Path
filepath = Path('../datasets/data_prepared.csv')
filepath.parent.mkdir(parents = True, exist_ok = True)
X_ready.to_csv(filepath)
```

```python
## разделение датасета на train и test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split (X_ready, Y, test_size = 0.2, random_state = 21)   ## random_state = 21 - задает начальное значение псевдослучайных чисел, чтобы они всегда были одинаковые
print('Размерность обучающей выборки X_train:', X_train.shape)
```

## Задание 7
Обучи логистическую регрессию с параметрами: `solver='liblinear'`, `fit_intercept=False`, `penalty='l1'`, `random_state=21`.
Посчитай `accuracy` модели на тестовой выборке, сравнив предсказанные значения с реальными.
Выведи значение `accuracy` модели на тестовой выборке.

```python
from sklearn. linear_model import LogisticRegression
log_regression = LogisticRegression(solver='liblinear', fit_intercept=False, penalty='l1', random_state=21)  ## создали экземпляр класса с начальными значениями
clf = log_regression.fit(X_train, Y_train) ## размещаем данные в нашей модели
```

```python
Y_pred = log_regression.predict(X_test) ## начинаем моделировать - прогноз данных на основе обученной модели
```

```python
from sklearn import metrics
print('accuracy модели на тестовой выборке =', metrics.accuracy_score (Y_test,Y_pred))
```

## Задание 9
Построй график, на котором видны топ-10 самых важных факторов по мнению модели в абсолютном значении.

```python
attribute_lst = X_ready.columns.tolist() # список признаков
attribute_df = pd.DataFrame(attribute_lst, columns=['attribute_name']) # DataFrame из признаков
coef_df = pd.DataFrame(np.abs(clf.coef_).T, columns = ['attribute_meaning']) # DataFrame из коэффициентов
result_df = pd.concat([attribute_df, coef_df], axis = 1)  # Объединяем DataFrame
result_df
```

```python
data = result_df.sort_values('attribute_meaning', ascending=False)[:10] #Срез из 10ти наибольших значений ascending=False - первые 10, scending=True - последние 10,
data
```

```python
import seaborn as sns
ax = sns.barplot(x=data['attribute_meaning'], y =data['attribute_name'], palette = 'colorblind' )
props = {
    'title': 'Топ-10 самых важных факторов по мнению модели в абсолютном значении',
    'xlabel': 'attribute_meaning',
    'ylabel': 'attribute_name',
}
ax.set(**props)
plt.xticks(rotation=30, ha='right', fontsize = 8)
plt.yticks(rotation=30, ha='right', fontsize = 8)
plt.xlabel('Абсолютные значения коэффициентов', fontsize = 8)
plt.xlabel('Названия признаков', fontsize = 8)
```
